```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")  # Get full history

# Filter data up to June 2021
tesla_data = tesla_data[tesla_data.index <= "2021-06-30"]

# Get historical financials (Revenue)
tesla_financials = tesla.financials.T
tesla_revenue = tesla_financials["Total Revenue"]

# Function to plot stock data
def make_stock_price_graph(stock_data, title="Tesla Stock Price (Up to June 2021)"):
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data["Close"], label="Closing Price", color="blue")
    plt.xlabel("Date")
    plt.ylabel("Stock Price (USD)")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

# Function to plot revenue data
def make_revenue_graph(revenue_data, title="Tesla Revenue (Up to June 2021)"):
    plt.figure(figsize=(12, 6))
    plt.plot(revenue_data.index, revenue_data, label="Revenue", color="green")
    plt.xlabel("Date")
    plt.ylabel("Revenue (USD)")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

# Plot stock price graph
make_stock_price_graph(tesla_data, title="Tesla Stock Price (Up to June 2021)")

# Plot revenue graph
make_revenue_graph(tesla_revenue, title="Tesla Revenue (Up to June 2021)")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Step 1: Download GameStop stock data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")  # Get full history

# Step 2: Filter data up to June 2021
gme_data = gme_data[gme_data.index <= "2021-06-30"]

# Step 3: Get GameStop's historical revenue data
gme_financials = gme.financials
revenue_data = gme_financials.loc['Total Revenue']

# Step 4: Define function to plot stock price and revenue data
def make_graph(stock_data, revenue_data, title="GameStop Stock Price and Revenue (Up to June 2021)"):
    plt.figure(figsize=(12, 6))

    # Plot Stock Price
    plt.subplot(2, 1, 1)  # Plotting in two subplots (1st subplot for stock price)
    plt.plot(stock_data.index, stock_data["Close"], label="Closing Price", color="red")
    plt.xlabel("Date")
    plt.ylabel("Stock Price (USD)")
    plt.title("GameStop Stock Price (Up to June 2021)")
    plt.legend()
    plt.grid()

    # Plot Revenue Data
    plt.subplot(2, 1, 2)  # 2nd subplot for revenue data
    plt.plot(revenue_data.index, revenue_data.values, label="Total Revenue", color="blue")
    plt.xlabel("Date")
    plt.ylabel("Revenue (USD)")
    plt.title("GameStop Total Revenue")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

# Step 5: Call function with GameStop data
make_graph(gme_data, revenue_data, title="GameStop Stock Price and Revenue (Up to June 2021)")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Step 1: Define the URL
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Step 2: Send a request to the website with a user-agent header
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"  # Example user-agent
}
response = requests.get(url, headers=headers)
response.raise_for_status()  # Raise an exception for bad status codes

# Step 3: Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Step 4: Find the table containing the revenue data
# (Inspect the website's HTML to identify the table's structure)
table = soup.find("table", {"class": "historical_data_table table"})  # Replace with the actual table class or attributes

# Step 5: Extract data from the table
# (Use BeautifulSoup's methods like find_all to extract rows and columns)
data = []
for row in table.find_all("tr")[1:]:  # Skip the header row
    cols = row.find_all("td")
    date = cols[0].text.strip()
    revenue = cols[1].text.strip().replace("$", "").replace(",", "")
    data.append([date, revenue])

# Step 6: Create a DataFrame
gme_revenue = pd.DataFrame(data, columns=["Date", "Revenue"])

# Step 7: Display the last 5 rows
print(gme_revenue.tail())
```
